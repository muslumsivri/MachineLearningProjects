## Defining train and test data set

```python
df.head()
```

```python
X = df.drop(["id","diagnosis"],axis=1)
X
```

```python
y = df["diagnosis"]
y
```

## Let's split data

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

## Now we have to scale our data

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

## Apply KNN Algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
knn = KNeighborsClassifier()
knn.get_params().keys()
```

```python
param_grid = {"n_neighbors":[3,5,7,9,15,25,35]}
grid = GridSearchCV(knn,param_grid,cv=5,scoring="accuracy")
```

```python
grid.fit(scaled_X_train,y_train)
```

```python
grid.best_params_
```

## -->best number of neighbours is 5 so we need to take it as a parameter

```python
final_model = KNeighborsClassifier(n_neighbors=5)
```

```python
final_model.fit(scaled_X_train,y_train)
```

```python
y_pred = final_model.predict(scaled_X_test)
```

## Validation

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

## As you can see Accuracy is 0.97 ; it is an acceptable value
